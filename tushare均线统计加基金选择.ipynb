```python
import tushare as ts
import pandas as pd
from datetime import datetime
import time
import re
from pandas.core.frame import DataFrame
today = str(datetime.now().date())
start = datetime.now()
print(start)
# 一分钟统计一百只股票
ts.set_token('b2c3c2f05cc2aa6dcf13f02e0110ffb7d968ceecfbfc446977c02850')
pro=ts.pro_api()  #基本上这三步肯定是需要的
df1 = pro.stock_basic(exchange='', list_status='L', fields='ts_code,symbol,name,area,industry,list_date')

list11=[]
list22=[]
for i in df1["ts_code"]:
    list11.append(i)
for i in df1["symbol"]:
    list22.append(i)

zd=dict(zip(list11,list22))

df2=pd.read_excel('2020年12月 基金前十大重仓股，成交额大于1千....xls',sheet_name=0,)
b=set(re.findall("[0-9]{6}",str(df2["基金十大重仓股票"].tolist()),))
list33=list(b)

list5=[]
list10=[]
list60=[]
list60j=[]
list250=[]
list250j=[]
codes=[]
for k,v in zd.items():
    if v in list33:
        codes.append(k)
print(len(codes))

for code in codes:
    
    try:
        df = pro.daily(ts_code=code, start_date='20150101',)
        df=df.set_index('trade_date')
        df_close = df['close']
        x = pd.DataFrame(df_close)
        #y = x.sort_index(axis='0', ascending=True)
        y= x.sort_values(by='trade_date', ascending=True)
        stockDate5 = y.rolling(5).mean()
        stockDate10 = y.rolling(10).mean()
        stockDate60 = y.rolling(60).mean()
        stockDate250 = y.rolling(250).mean()
        value5 = float('%.2f' % stockDate5.iloc[-1])
        value10 = float('%.2f' % stockDate10.iloc[-1])
        value60 = float('%.2f' % stockDate60.iloc[-1])
        value250 = float('%.2f' % stockDate250.iloc[-1])
        value60j = float('%.2f' % stockDate60.iloc[-31])
        value250j = float('%.2f' % stockDate250.iloc[-125])
        close = float('%.2f' % y.iloc[-1])
        close60j = float('%.2f' % y.iloc[-31])
        close250j = float('%.2f' % y.iloc[-125])
        print("正在计算"+code)

        if close > value5 :
            list5.append(code)
        if close > value10 :
            list10.append(code)
        if close > value60 :
            list60.append(code)
        if close60j > value60j :
            list60j.append(code)
        if close > value250 :
            list250.append(code)
        if close250j > value250j :
            list250j.append(code)
    except:
        pass

a=set(re.findall("[0-9]{6}",str(list60)))
c=set(re.findall("[0-9]{6}",str(list60j)))
d=set(re.findall("[0-9]{6}",str(list10)))
e=set(re.findall("[0-9]{6}",str(list5)))
f=set(re.findall("[0-9]{6}",str(list250)))
g=set(re.findall("[0-9]{6}",str(list250j)))

df=pd.read_excel('2020年12月 基金前十大重仓股，成交额大于1千....xls',sheet_name=0,)
list9=[]
list6=[]
list7=[]
list8=[]
list5=[]
list4=[]
for i in df["基金十大重仓股票"]:
    b=set(re.findall("[0-9]{6}",str(i),))
    num1=len(a&b)
    num2=len(c&b)
    num3=len(d&b)
    num4=len(e&b)
    num5=len(f&b)
    num6=len(g&b)
    list9.append(num1)
    list6.append(num2)
    list7.append(num3)
    list8.append(num4)
    list5.append(num5)
    list4.append(num6)

df['60次数']=DataFrame(list9)
df['60次数j']=DataFrame(list6)
df['10次数']=DataFrame(list7)
df['5次数']=DataFrame(list8)
df['250次数']=DataFrame(list5)
df['250次数j']=DataFrame(list4)
df=df[["250次数","250次数j","60次数","60次数j","10次数","5次数","基金代码","基金简称",]]
df['250次数-250次数j'] = df['250次数'] - df['250次数j'] 
df['60次数-60次数j'] = df['60次数'] - df['60次数j'] 
df['和250差和60差']=df['250次数-250次数j']+ df['60次数-60次数j']
df['60次数-10次数'] = df['60次数'] - df['10次数'] 
df['60次数-5次数'] = df['60次数'] - df['5次数'] 
df['综合']=df['60次数-60次数j']+df['60次数-10次数']+df['60次数-5次数']
df['综合1']=df['60次数-10次数']+df['60次数-5次数']
# 筛选条件如下
df=df[(df['250次数-250次数j']>=1)|(df['60次数-60次数j']>=1)]
df=df[df['60次数']>=5]
df=df[(df['10次数']<=3)&(df['5次数']<=2)]

df=df.sort_values(by=["综合"],ascending=False)#ascending=False
df=df[["250次数","60次数","10次数","5次数","基金代码","综合","综合1","基金简称",]]

df10=pd.read_excel('基金次数总数统计结果.xls',)
df10=df10[["基金代码","股票次数总和","基金十大重仓股票"]]
df11=pd.merge(df,df10,on=["基金代码"])
df11.to_excel(today+"基金统计结果全.xls")

end = datetime.now()
print(end-start)
print(df11)
```
